```python
import fitz  # PyMuPDF
import os

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text
```

```python
# Example: process all PDFs in upload folder
pdf_texts = []
for filename in uploaded:
    if filename.endswith(".pdf"):
        content = extract_text_from_pdf(f"/content/drive/My Drive/IncludedPapers/{filename}")
        pdf_texts.append({
            "filename": filename,
            "content": content
        })
```

```python
pdf_texts[0]
```

```python
import pandas as pd
import textwrap

chunks = []
for pdf in pdf_texts:
    chunks += [
        {"source": pdf["filename"], "content": chunk}
        for chunk in textwrap.wrap(pdf["content"], width=512)
    ]

df = pd.DataFrame(chunks)

```

```python
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
from sentence_transformers import SentenceTransformer

retriever = SentenceTransformer(
    # 'sentence-transformers/multi-qa-mpnet-base-dot-v1'
    "flax-sentence-embeddings/all_datasets_v3_mpnet-base",
    device=device
    )

retriever
```

```python
df["embedding"]= df["content"].apply(lambda x: retriever.encode(x).tolist())
```

```python
# !pip install -U pinecone-client

from pinecone import Pinecone, ServerlessSpec, CloudProvider, AwsRegion, Metric

API_KEY = API_KEY

pc = Pinecone(api_key=API_KEY, environment="us-east-1")

index_name = "abstractive-question-answering"

pc.delete_index("abstractive-question-answering")

if index_name not in pc.list_indexes():
    pc.create_index(
        name=index_name,
        dimension=768,
        metric=Metric.COSINE,
        spec=ServerlessSpec(
            cloud=CloudProvider.AWS,
            region=AwsRegion.US_EAST_1
        )
    )

index = pc.Index(index_name)
```

```python
index = pc.Index(index_name)
```

```python
from tqdm.auto import tqdm

batch_size = 64
for i in tqdm(range(0, len(df), batch_size)):
    batch = df.iloc[i:i+batch_size]
    ids = [f"{i + j}" for j in range(len(batch))]
    vectors = list(zip(ids, batch["embedding"], batch[["content", "source"]].to_dict("records")))
    index.upsert(vectors=vectors)
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration

tokenizer = BartTokenizer.from_pretrained("vblagoje/bart_lfqa")
generator = BartForConditionalGeneration.from_pretrained("vblagoje/bart_lfqa")
```

```python
def ask_question(query, top_k=3):
    # Embed question
    query_vector = retriever.encode(query).tolist()
    results = index.query(vector=query_vector, top_k=top_k, include_metadata=True)

    # Format input for generator
    context = " ".join([f"<P> {m['metadata']['content']}" for m in results["matches"]])
    formatted_input = f"question: {query} context: {context}"

    # Generate answer
    input_ids = tokenizer.encode(formatted_input, return_tensors="pt")
    output_ids = generator.generate(input_ids, max_new_tokens=64)
    answer = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    return answer
```

```python
ask_question("authors the Haplotypes of TNFα/β Genes Associated with Sex-Specific Paranoid Schizophrenic Risk in Tunisian Population ")
```
